# Using the UNSW QRS detector as a standalone tool

This notebook provides a demonstration of how to use the UNSW QRS detector as a standalone tool.

## Setup

Import required packages

```python
import numpy as np
import sys
import os
import matplotlib.pyplot as plt
from scipy.datasets import electrocardiogram # for loading sample ECG signal
```

Import UNSW QRS detector

```python
sys.path.append(os.path.join(os.path.dirname(__file__), 'src') if '__file__' in globals() else './src')
from UNSW_QRSDetector import UNSW_QRSDetector
```

Import sample ECG data

```python
rawecg = electrocardiogram() # load sample ECG signal
fs = 360 # in Hz, as specified at in scipy documentation
```

## Detect QRS complexes

```python
unswdata = UNSW_QRSDetector(rawecg,fs) # perform QRS detection
qrs_indices = unswdata['qrs'] # extract QRS indices
```

## Plot ECG and detected QRS complexes

Create time vector

```python
time = np.arange(len(rawecg)) / fs
```

Plot

```python
plt.plot(time, rawecg)
plt.scatter(time[qrs_indices], rawecg[qrs_indices], color='red')
plt.xlabel('Time (s)')
plt.ylabel('ECG (mV)')
plt.title('ECG Signal and Detected QRS Complexes')
plt.xlim([0, 20])
plt.show()
```
