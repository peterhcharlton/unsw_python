# Using the UNSW QRS detector in Neurokit2

This notebook provides a demonstration of how to use the UNSW QRS detector implemented in Neurokit2.

## Setup

Import required packages

```python
import neurokit2 as nk
```

Simulate example ECG signal

```python
sampling_rate = 360
duration = 10
ecg = nk.ecg_simulate(duration=duration, sampling_rate=sampling_rate)
```

## Detect QRS complexes

Detect QRS complexes using the UNSW QRS detector implemented in Neurokit (named "Khamis2016"), and plot

```python
signals, info = nk.ecg_peaks(ecg, sampling_rate=250, correct_artifacts=True, show=True, method="khamis2016")
```
